```python
import os
import random
import shutil
from pathlib import Path
import json
```

```python
random_seed = 2025
random.seed(random_seed)
NUM_CLASS = 10
```

```python
ABS_SRC_PATH = "F:/Data/"
ABS_TARGET_PATH = "../Data/GenImage/"
generator_names = ["adm","bgan","glide","midj","sd_14","sd_15","vqdm","wukong"]
```

```python
def map1k21k(idx_1k):
    names_21k = []
    idx_21k = []
    names_1k = []
    with open("../Data/GenImage/imagenet21k_wordnet_ids.txt", "r", encoding="utf-8") as f:
        for idx, line in enumerate(f):
            line = line.strip()  
            idx_21k.append(idx+1)
            names_21k.append(line)
    with open("../Data/GenImage/map21k21k.txt", "r", encoding="utf-8") as f:
        for idx, line in enumerate(f):
            line = line.strip()
            names_1k.append(line)
            assert int(line) != 9206 and int(line) != 15028
    print(len(names_21k),len(names_1k))
    corresponding_names = [names_21k[int(names_1k[int(idx_1k[i])])] for i in range(len(idx_1k))]
    return corresponding_names
```

```python
all_classes = [f"{j:03d}" for j in range(1000)]
```

```python
classes_names = map1k21k(all_classes)
classes_dict = {"1k_idx":all_classes, "21k_idx":classes_names}
with open("class.json", 'w') as f:
    json.dump(classes_dict, f)
```

```python
classes_idx = random.sample(all_classes, NUM_CLASS)
classes_names = map1k21k(classes_idx)
classes_dict = {"1k_idx":classes_idx, "21k_idx":classes_names}
with open("classes.json", 'w') as f:
    json.dump(classes_dict, f)
```

```python
def nature_extractor(nature_dir, nature_target_dir,corresponding_names,images_per_class):
    Path(nature_target_dir).mkdir(parents=True, exist_ok=True)
    total_copied = 0
    for class_id in corresponding_names:
        pattern = f"{class_id}_*"
        class_files = list(Path(nature_dir).glob(pattern))
        # real_images = list(Path(nature_dir).glob("*"))
        print(f"Class{class_id} has {len(class_files)} pictures.")
        if not class_files:
            print(f"Error: Class{class_id} has no pictures.")
            continue
        real_images = random.sample(class_files, images_per_class)
        for img_path in real_images:
            target_path = Path(nature_target_dir) / img_path.name
            shutil.copy2(img_path, target_path)
            total_copied += 1
    print(f"{total_copied} pictures have been copied into {nature_target_dir}.")  


def ai_extractor_1(ai_dir, ai_target_dir, classes, images_per_class):
    Path(ai_target_dir).mkdir(parents=True, exist_ok=True)
    total_copied = 0
    for class_id in classes:
        pattern = f"{class_id}_biggan_*.png"
        class_files = list(Path(ai_dir).glob(pattern))
        print(f"Class{class_id} has {len(class_files)} pictures.")
        if not class_files:
            print(f"Error: Class{class_id} has no pictures.")
            continue
        images = random.sample(class_files, min(images_per_class, len(class_files)))
        for img_path in images:
            target_path = Path(ai_target_dir) / img_path.name
            shutil.copy2(img_path, target_path)
            total_copied += 1
    print(f"{total_copied} pictures have been copied into {ai_target_dir}.")

def ai_extractor_2(ai_dir, ai_target_dir, classes, images_per_class):
    Path(ai_target_dir).mkdir(parents=True, exist_ok=True)
    total_copied = 0
    for class_id in classes:
        pattern = f"{class_id}_sdv5_*.png"
        class_files = list(Path(ai_dir).glob(pattern))
        print(f"Class{class_id} has {len(class_files)} pictures.")
        if not class_files:
            print(f"Error: Class{class_id} has no pictures.")
            continue
        images = random.sample(class_files, min(images_per_class, len(class_files)))
        for img_path in images:
            target_path = Path(ai_target_dir) / img_path.name
            shutil.copy2(img_path, target_path)
            total_copied += 1
    print(f"{total_copied} pictures have been copied into {ai_target_dir}.")

def ai_extractor_3(ai_dir, ai_target_dir, classes, images_per_class):
    Path(ai_target_dir).mkdir(parents=True, exist_ok=True)
    total_copied = 0
    for class_id in classes:
        pattern = f"{int(class_id)}_midjourney_*.png"
        class_files = list(Path(ai_dir).glob(pattern))
        print(f"Class{class_id} has {len(class_files)} pictures.")
        if not class_files:
            print(f"Error: Class{class_id} has no pictures.")
            continue
        images = random.sample(class_files, min(images_per_class, len(class_files)))
        for img_path in images:
            target_path = Path(ai_target_dir) / img_path.name
            shutil.copy2(img_path, target_path)
            total_copied += 1
    print(f"{total_copied} pictures have been copied into {ai_target_dir}.")
```

```python
def extractor(classes, global_name, corresponding_names, images_per_class):
    assert global_name in generator_names
    nature_target_dir = ABS_TARGET_PATH + "nature"
    Path(nature_target_dir).mkdir(parents=True, exist_ok=True)
    match global_name:
        case "glide":
            source_dir = ABS_SRC_PATH + "imagenet_glide/train/ai"
            nature_dir = ABS_SRC_PATH + "imagenet_glide/train/nature"
            target_dir = ABS_TARGET_PATH + "glide"
            Path(target_dir).mkdir(parents=True, exist_ok=True)
            total_copied = 0
            for class_id in classes:
                pattern = f"*_{class_id}_glide_*.png"
                class_files = list(Path(source_dir).glob(pattern))
                print(f"Class{class_id} has {len(class_files)} pictures.")
                if not class_files:
                    print(f"Error: Class{class_id} has no pictures.")
                    continue
                images = random.sample(class_files, min(images_per_class, len(class_files)))
                for img_path in images:
                    target_path = Path(target_dir) / img_path.name
                    shutil.copy2(img_path, target_path)
                    total_copied += 1
            print(f"{total_copied} pictures have been copied into {target_dir}.")
            nature_extractor(nature_dir, nature_target_dir, corresponding_names[0],2)
        case "adm":
            source_dir = ABS_SRC_PATH + "imagenet_adm/train/ai"
            nature_dir = ABS_SRC_PATH + "imagenet_adm/train/nature"
            target_dir = ABS_TARGET_PATH + "adm"
            Path(target_dir).mkdir(parents=True, exist_ok=True)
            total_copied = 0
            for class_id in classes:
                pattern = f"{int(class_id)}_adm_*.png"
                class_files = list(Path(source_dir).glob(pattern))
                print(f"Class{class_id} has {len(class_files)} pictures.")
                if not class_files:
                    print(f"Error: Class{class_id} has no pictures.")
                    continue
                images = random.sample(class_files, min(images_per_class, len(class_files)))
                for img_path in images:
                    target_path = Path(target_dir) / img_path.name
                    shutil.copy2(img_path, target_path)
                    total_copied += 1
            print(f"{total_copied} pictures have been copied into {target_dir}.")
            nature_extractor(nature_dir, nature_target_dir, corresponding_names[1],2)
        case "bgan":
            source_dir = ABS_SRC_PATH + "imagenet_bgan/train/ai"
            nature_dir = ABS_SRC_PATH + "imagenet_bgan/train/nature"
            target_dir = ABS_TARGET_PATH + "bgan"
            Path(target_dir).mkdir(parents=True, exist_ok=True)
            total_copied = 0
            for class_id in classes:
                pattern = f"{class_id}_biggan_*.png"
                class_files = list(Path(source_dir).glob(pattern))
                print(f"Class{class_id} has {len(class_files)} pictures.")
                if not class_files:
                    print(f"Error: Class{class_id} has no pictures.")
                    continue
                images = random.sample(class_files, min(images_per_class, len(class_files)))
                for img_path in images:
                    target_path = Path(target_dir) / img_path.name
                    shutil.copy2(img_path, target_path)
                    total_copied += 1
            print(f"{total_copied} pictures have been copied into {target_dir}.")
            nature_extractor(nature_dir, nature_target_dir, corresponding_names[2],2)
        case "midj":
            source_dir = ABS_SRC_PATH + "imagenet_midj/train/ai"
            nature_dir = ABS_SRC_PATH + "imagenet_midj/train/nature"
            target_dir = ABS_TARGET_PATH + "midj"
            Path(target_dir).mkdir(parents=True, exist_ok=True)
            total_copied = 0
            for class_id in classes:
                pattern = f"{int(class_id)}_midjourney_*.png"
                class_files = list(Path(source_dir).glob(pattern))
                print(f"Class{class_id} has {len(class_files)} pictures.")
                if not class_files:
                    print(f"Error: Class{class_id} has no pictures.")
                    continue
                images = random.sample(class_files, min(images_per_class, len(class_files)))
                for img_path in images:
                    target_path = Path(target_dir) / img_path.name
                    shutil.copy2(img_path, target_path)
                    total_copied += 1
            print(f"{total_copied} pictures have been copied into {target_dir}.")
            nature_extractor(nature_dir, nature_target_dir, corresponding_names[3],2)
        case "sd_14":
            source_dir = ABS_SRC_PATH + "imagenet_sd_14/train/ai"
            nature_dir = ABS_SRC_PATH + "imagenet_sd_14/train/nature"
            target_dir = ABS_TARGET_PATH + "sd_14"
            Path(target_dir).mkdir(parents=True, exist_ok=True)
            total_copied = 0
            for class_id in classes:
                pattern = f"{class_id}_sdv4_*.png"
                class_files = list(Path(source_dir).glob(pattern))
                print(f"Class{class_id} has {len(class_files)} pictures.")
                if not class_files:
                    print(f"Error: Class{class_id} has no pictures.")
                    continue
                images = random.sample(class_files, min(images_per_class, len(class_files)))
                for img_path in images:
                    target_path = Path(target_dir) / img_path.name
                    shutil.copy2(img_path, target_path)
                    total_copied += 1
            print(f"{total_copied} pictures have been copied into {target_dir}.")
            nature_extractor(nature_dir, nature_target_dir, corresponding_names[4],2)
        case _:
            print("No such generator files.")


    
```
